# Categorical Data- Bar charts & Pie charts

```python
data['Pclass'].value_counts()
```

```python
data['Pclass'].value_counts().plot(kind='bar')
```

```python
data['Pclass'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
a=data['Pclass'].value_counts()
```

```python
a
```

```python
b=((a/a.sum())*100).values
```

```python
b
```

```python
c=[]
cf=0
for i in b:
    cf=cf+i
    c.append(cf)
```

```python
c
```

```python
plt.plot(c,marker='D')
```
